```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Charger les donn√©es
df = pd.read_csv('air_quality_data.csv')

# Convertir la colonne timestamp en datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
# Distribution des niveaux de PM2.5 par ville
plt.figure(figsize=(10, 6))
sns.boxplot(x='city', y='pm25', data=df)
plt.title('Distribution des niveaux de PM2.5 par ville')
plt.xlabel('Ville')
plt.ylabel('PM2.5')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Figure_1.png')
```

```python
# Niveaux de PM2.5 au fil du temps par ville
df['timestamp'] = pd.to_datetime(df['timestamp'])
plt.figure(figsize=(10, 6))
for city in df['city'].unique():
    city_data = df[df['city'] == city]
    plt.plot(city_data['timestamp'], city_data['pm25'], label=city)
plt.title('Niveaux de PM2.5 au fil du temps par ville')
plt.xlabel('Temps')
plt.ylabel('PM2.5')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('Figure_2.png')
```

```python
# Matrices de corr√©lation des polluants par ville
pollutants = ['pm25', 'pm10', 'no2', 'so2', 'o3', 'co']
for city in df['city'].unique():
    city_data = df[df['city'] == city][pollutants]
    corr_matrix = city_data.corr()
    plt.figure(figsize=(10, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
    plt.title(f'Matrice de corr√©lation des polluants - {city}')
    plt.tight_layout()
    plt.savefig(f'Figure_{city}.png')
```

```python
# Matrice de corr√©lation des polluants par ville
for city in df['city'].unique():
    df_city = df[df['city'] == city]
    corr = df_city[['pm25', 'pm10', 'no2', 'so2', 'o3', 'co']].corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm')
    plt.title(f'Matrice de corr√©lation des polluants - {city}')
    plt.show()
```
